# Notebook 4 — Mobility KPI Forecasting
**Target KPI:** `handover_success_rate_percent`

| Model | Why |
|---|---|
| **LSTM** | Baseline — captures temporal handover patterns |
| **GRU** | Lighter — handover signals can be sparse; GRU avoids overfitting |
| **DNN** | Feedforward on flattened window — tests if current state alone predicts handover success |

```python
import math
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import joblib
import os
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# ─────────────────────────────────────────────
#  CONFIGURATION  —  edit these values freely
# ─────────────────────────────────────────────

DATA_PATH   = "../Data/Model_data.csv"
OUTPUT_DIR  = "../outputs/mobility"

TARGET_KPI  = "handover_success_rate_percent"
TOLERANCE   = 0.5       # ± 0.5 %

SEQ_LEN     = 12
TRAIN_RATIO = 0.8

HIDDEN_SIZE = 64
NUM_LAYERS  = 2
EPOCHS      = 10
BATCH_SIZE  = 2048
LR          = 1e-3

RNG_SEED = 42
DEVICE   = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
np.random.seed(RNG_SEED)
torch.manual_seed(RNG_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RNG_SEED)

os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"Device: {DEVICE}")
```

```python
print("Loading data...")
df = pd.read_csv(DATA_PATH, parse_dates=["timestamp"])
df = df.sort_values("timestamp").reset_index(drop=True)
print(f"Rows: {len(df):,} | Columns: {len(df.columns)}")
df.head(3)
```

```python
# Mobility features + lag columns
FEATURE_COLS = [
    "handover_success_rate_percent",
    "handover_interruption_time_ms",
    "one_way_latency_ms",
    "rtt_ms",
    "throughput_dl_mbps",
    "bler_percent",
    "packet_loss_percent",
]

df = df.dropna(subset=FEATURE_COLS).reset_index(drop=True)
print(f"Clean rows: {len(df):,}")
```

```python
print("\nBuilding sequences...")

X_raw = df[FEATURE_COLS].values
y_raw = df[TARGET_KPI].shift(-1).values
X_raw, y_raw = X_raw[:-1], y_raw[:-1]

X_all, y_all = [], []
for i in range(SEQ_LEN, len(X_raw)):
    X_all.append(X_raw[i - SEQ_LEN : i, :])
    y_all.append(y_raw[i])

X_seq = np.stack(X_all).astype(np.float32)
y_seq = np.array(y_all, dtype=np.float32).reshape(-1, 1)
print(f"Sequences: {len(X_seq):,} | X: {X_seq.shape} | y: {y_seq.shape}")
```

```python
train_n = int(len(X_seq) * TRAIN_RATIO)
X_train, X_test = X_seq[:train_n], X_seq[train_n:]
y_train, y_test = y_seq[:train_n], y_seq[train_n:]

scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train.reshape(-1, X_train.shape[2])).reshape(X_train.shape)
X_test_scaled  = scaler_X.transform(X_test.reshape(-1, X_test.shape[2])).reshape(X_test.shape)

scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train)
y_test_scaled  = scaler_y.transform(y_test)

X_train_t = torch.tensor(X_train_scaled, dtype=torch.float32, device=DEVICE)
X_test_t  = torch.tensor(X_test_scaled,  dtype=torch.float32, device=DEVICE)
y_train_t = torch.tensor(y_train_scaled, dtype=torch.float32, device=DEVICE)
y_test_t  = torch.tensor(y_test_scaled,  dtype=torch.float32, device=DEVICE)

# DNN uses a flattened view of the sequence
flat_size      = SEQ_LEN * X_train_t.shape[2]
X_train_flat_t = X_train_t.reshape(len(X_train_t), -1)
X_test_flat_t  = X_test_t.reshape(len(X_test_t), -1)

print(f"Train: {len(X_train_t):,} | Test: {len(X_test_t):,} | Features: {X_train_t.shape[2]} | Device: {DEVICE}")
print(f"DNN flat input size: {flat_size}")
```

```python
# ── MODEL 1: LSTM ──────────────────────────────────────────────────────────────
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size=64, num_layers=2, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True,
                            dropout=dropout if num_layers > 1 else 0.0)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])


# ── MODEL 2: GRU ───────────────────────────────────────────────────────────────
class GRUModel(nn.Module):
    def __init__(self, input_size, hidden_size=64, num_layers=2, dropout=0.2):
        super().__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True,
                          dropout=dropout if num_layers > 1 else 0.0)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])


# ── MODEL 3: DNN ───────────────────────────────────────────────────────────────
class DNNModel(nn.Module):
    def __init__(self, flat_input_size, dropout=0.2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(flat_input_size, 256), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(256, 128),             nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(128, 64),              nn.ReLU(),
            nn.Linear(64, 1),
        )
    def forward(self, x):
        return self.net(x)
```

```python
def get_batches(X, y, batch_size):
    for i in range(0, len(X), batch_size):
        yield X[i : i + batch_size], y[i : i + batch_size]


def train_and_evaluate(model, name, use_flat=False):
    print(f"\n{'─'*55}\n  Training: {name}\n{'─'*55}")
    Xtr = X_train_flat_t if use_flat else X_train_t
    Xte = X_test_flat_t  if use_flat else X_test_t

    optimizer = torch.optim.Adam(model.parameters(), lr=LR)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=2, factor=0.5)
    loss_fn   = nn.L1Loss()

    for epoch in range(1, EPOCHS + 1):
        model.train()
        train_losses = []
        for bx, by in get_batches(Xtr, y_train_t, BATCH_SIZE):
            optimizer.zero_grad()
            loss = loss_fn(model(bx), by)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())

        model.eval()
        with torch.no_grad():
            preds = scaler_y.inverse_transform(model(Xte).cpu().numpy())
            true  = scaler_y.inverse_transform(y_test_t.cpu().numpy())
            mae   = mean_absolute_error(true, preds)
            rmse  = math.sqrt(mean_squared_error(true, preds))

        mean_train = np.mean(train_losses)
        scheduler.step(mean_train)
        print(f"  Epoch {epoch}/{EPOCHS} | Train MAE: {mean_train:.4f} | Test MAE: {mae:.4f} | RMSE: {rmse:.4f} | LR: {optimizer.param_groups[0]['lr']:.6f}")

    acc = np.mean(np.abs(true - preds) <= TOLERANCE) * 100
    return {"name": name, "model": model, "MAE": mae, "RMSE": rmse, "Accuracy": acc,
            "preds": preds, "true": true, "use_flat": use_flat}
```

```python
input_size = X_train_t.shape[2]
results = []
results.append(train_and_evaluate(LSTMModel(input_size, HIDDEN_SIZE, NUM_LAYERS).to(DEVICE), "LSTM"))
results.append(train_and_evaluate(GRUModel(input_size, HIDDEN_SIZE, NUM_LAYERS).to(DEVICE),  "GRU"))
results.append(train_and_evaluate(DNNModel(flat_size).to(DEVICE), "DNN", use_flat=True))
```

```python
print("\n" + "═"*62)
print(f"  MODEL COMPARISON  —  Target: {TARGET_KPI}")
print("═"*62)
print(f"  {'Model':<10} {'MAE':>10} {'RMSE':>10} {'Acc (±{:.1f})'.format(TOLERANCE):>14}")
print("  " + "─"*56)
best = min(results, key=lambda r: r["MAE"])
for r in results:
    marker = " ← BEST" if r["name"] == best["name"] else ""
    print(f"  {r['name']:<10} {r['MAE']:>10.4f} {r['RMSE']:>10.4f} {r['Accuracy']:>13.2f}%{marker}")
print("═"*62)
```

```python
# === LOAD DASHBOARD DATA & BUILD SEQUENCES PER CELL ===
DASHBOARD_PATH = "../Data/Dashboard_data.csv"

print("Loading dashboard data...")
df_dash = pd.read_csv(DASHBOARD_PATH, parse_dates=["timestamp"])
df_dash = df_dash.sort_values(["cell_id", "timestamp"]).reset_index(drop=True)
df_dash = df_dash.dropna(subset=FEATURE_COLS).reset_index(drop=True)
print(f"Rows: {len(df_dash):,} | Cells: {df_dash['cell_id'].nunique()}")

# Build sequences per cell to avoid cross-cell contamination
X_dash_list, meta_list = [], []

for cell_id, group in df_dash.groupby("cell_id", sort=True):
    group = group.sort_values("timestamp").reset_index(drop=True)
    X_raw_c = group[FEATURE_COLS].values
    for i in range(SEQ_LEN, len(X_raw_c)):
        X_dash_list.append(X_raw_c[i - SEQ_LEN : i, :])
        meta_list.append({
            "timestamp":    group.iloc[i]["timestamp"],
            "cell_id":      cell_id,
            "ue_id":        group.iloc[i]["ue_id"],
            "slice_type":   group.iloc[i]["slice_type"],
            "actual_value": float(X_raw_c[i, FEATURE_COLS.index(TARGET_KPI)]),
        })

X_dash = np.stack(X_dash_list).astype(np.float32)  # (N, SEQ_LEN, F)
print(f"Dashboard sequences: {len(X_dash):,} | Shape: {X_dash.shape}")

# Helper: scale + optionally flatten a batch of raw sequences
def prepare_input(seq_raw_batch, use_flat=False):
    B = seq_raw_batch.shape[0]
    scaled = scaler_X.transform(
        seq_raw_batch.reshape(-1, seq_raw_batch.shape[2])
    ).reshape(seq_raw_batch.shape).astype(np.float32)
    t = torch.tensor(scaled, device=DEVICE)
    if use_flat:
        t = t.reshape(B, -1)
    return t

# Quick sanity check — show 4 sample predictions on dashboard sequences
best_model = best["model"]
best_model.eval()
use_flat = best["use_flat"]

sample_indices = [0, len(X_dash)//3, 2*len(X_dash)//3, len(X_dash)-1]
rows = []
with torch.no_grad():
    for idx in sample_indices:
        seq_raw = X_dash[idx:idx+1]  # (1, SEQ_LEN, F)
        seq_t   = prepare_input(seq_raw, use_flat)
        pred_scaled = best_model(seq_t).cpu().numpy()
        pred_val    = scaler_y.inverse_transform(pred_scaled)[0, 0]
        true_val    = meta_list[idx]["actual_value"]
        error       = abs(pred_val - true_val)
        rows.append({
            "Sample index":              idx,
            "cell_id":                   meta_list[idx]["cell_id"],
            f"Last known {TARGET_KPI}": round(X_dash[idx, -1, FEATURE_COLS.index(TARGET_KPI)].item(), 4),
            "Predicted (t+1)":          round(float(pred_val), 4),
            "Actual (t+1)":             round(true_val, 4),
            "Abs Error":                round(error, 4),
            f"Within ±{TOLERANCE}": "✓" if error <= TOLERANCE else "✗",
        })

print(f"\nBest model : {best['name']} (use_flat={use_flat})")
print(f"Target KPI : {TARGET_KPI}")
print(f"Tolerance  : ±{TOLERANCE} %")
print()
display(pd.DataFrame(rows))
```

```python
# === PREDICT ON ALL DASHBOARD DATA & EXPORT ===
best_model = best["model"]
best_model.eval()
use_flat = best["use_flat"]

PRED_BATCH = 4096
predictions = []
mae = best["MAE"]

with torch.no_grad():
    for start in range(0, len(X_dash), PRED_BATCH):
        batch      = X_dash[start : start + PRED_BATCH]     # (B, SEQ_LEN, F)
        t          = prepare_input(batch, use_flat)
        preds_vals = scaler_y.inverse_transform(
            best_model(t).cpu().numpy()
        ).flatten()

        for j, pred_val in enumerate(preds_vals):
            meta = meta_list[start + j]
            predictions.append({
                "timestamp":        str(meta["timestamp"]),
                "cell_id":          meta["cell_id"],
                "ue_id":            meta["ue_id"],
                "slice_type":       meta["slice_type"],
                "kpi_name":         TARGET_KPI,
                "actual_value":     round(meta["actual_value"], 4),
                "predicted_value":  round(float(pred_val), 4),
                "confidence_upper": round(float(pred_val) + 1.5 * mae, 4),
                "confidence_lower": round(max(0.0, float(pred_val) - 1.5 * mae), 4),
                "model_name":       best["name"],
            })

df_pred = pd.DataFrame(predictions)
os.makedirs(OUTPUT_DIR, exist_ok=True)
out_path = f"{OUTPUT_DIR}/{TARGET_KPI}_predictions.csv"
df_pred.to_csv(out_path, index=False)
print(f"Exported {len(df_pred):,} predictions → {out_path}")
print(f"Cells covered: {df_pred['cell_id'].nunique()}")
df_pred.head()
```
